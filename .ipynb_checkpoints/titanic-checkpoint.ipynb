```python
#4
# dont worrry about the error, just compiled the cabin string values into a list, and count the nan values
cabins = []
a = 0
for i in X.Cabin:
    if str(i) == 'nan':
        a += 1
    cabins.append(i)

print(a)
print(cabins)
```

```python
#5
floors = []
rooms = []
decks = []

for i in range(len(cabins)):
    enum = cabins[i].split("/")
    floors.append(ord(enum[0]))
    rooms.append(enum[1])
    if enum[2] == 'S':
        decks.append(0)
    else:
        decks.append(1)

X['Floor'] = floors
X['Room'] = rooms
X['Deck'] = decks

X = X.drop('Cabin', axis=1)
X.head()
```

```python
# Convert the CryoSleep: bool and VIP:bool to 0 or 1
Cryo = []
VIP = []

crval = X['CryoSleep'].tolist()
vval = X['VIP'].tolist()

for i in range(len(X['CryoSleep'])):
    if crval[i]:
        Cryo.append(1)
    else:
        Cryo.append(0)

        
for i in range(len(X['VIP'])):
    if vval[i]:
        VIP.append(1)
    else:
        VIP.append(0)

X['CryoSleep'] = Cryo
X['VIP'] = VIP
```

```python
# Standadizing the room values
rooms_std = []

for val in X.Room:
    if val == 0:
        rooms_std.append(0)
    else:
        rooms_std.append(int(val) / int(len(X.Room)))

X.Room = rooms_std
```

```python
# Standadizing the luxury values
Service = []
Court = []
Mall = []
Spa = []
Deck = []

headers = ['FoodCourt', 'ShoppingMall']

for val in X.RoomService:
    if val == 0:
        Service.append(0)
    else:
        Service.append(1 / int(val))

for val in X.FoodCourt:
    if val == 0:
        Court.append(0)
    else:
        Court.append(1 / int(val))

for val in X.ShoppingMall:
    if val == 0:
        Mall.append(0)
    else:
        Mall.append(1 / int(val))

for val in X.Spa:
    if val == 0:
        Spa.append(0)
    else:
        Spa.append(1 / int(val))

for val in X.VRDeck:
    if val == 0:
        Deck.append(0)
    else:
        Deck.append(1 / int(val))

X.FoodCourt = Court
X.ShoppingMall = Mall
X.Spa = Spa
X.VRDeck = Deck
```

```python
# standardize the floor value, 8 possible values so in 8ths
key = {65:0, 66:1, 66:2, 67:3, 68:4, 69:5, 70:6, 71:7, 84:8}
newFloor_val = []

for val in X.Floor:
    if val == 65:
        newFloor_val = newFloor_val + [0]
    else:
        newFloor_val = newFloor_val + [int(key[val]) / 8]

X.Floor = newFloor_val
```
